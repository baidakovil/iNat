```python
import pandas as pd
import requests
pd.set_option('display.max_columns', None)
```

```python
usual_columns = ['taxon_id','scientific_name','common_name','taxon_species_name']
preusual_columns = ['taxon_subclass_name', 'taxon_superorder_name',
       'taxon_order_name', 'taxon_suborder_name', 'taxon_superfamily_name',
       'taxon_family_name', 'taxon_subfamily_name', 'taxon_supertribe_name',
       'taxon_tribe_name', 'taxon_subtribe_name', 'taxon_genus_name',
       'taxon_genushybrid_name']
subusual_columns = ['taxon_hybrid_name','taxon_subspecies_name', 'taxon_variety_name', 'taxon_form_name']
all_columns = usual_columns + preusual_columns + subusual_columns
```

```python
observations_path = 'data/observations-285432.csv'

def prepare_df(observations_path):
    
    df_full = pd.read_csv(observations_path)
    df = df_full[all_columns].copy()

    return df, df_full

df, df_full = prepare_df(observations_path)
df_full.columns
```

```python
# наблюдения с точностью до подвида
df_subspecies = df[df['taxon_subspecies_name'].notnull()].copy()

subusual_to_remove = subusual_columns.copy()
subusual_to_remove.remove('taxon_subspecies_name')

df_subspecies.drop(preusual_columns, axis=1, inplace=True)
df_subspecies.drop(subusual_to_remove, axis=1, inplace=True)

df_subspecies
```

```python
# наблюдения с точностью до вида
df_species = df[df['taxon_species_name'].notnull() & df['taxon_subspecies_name'].isnull()].copy()
df_species.drop(preusual_columns, axis=1, inplace=True)
df_species.drop(subusual_columns, axis=1, inplace=True)
df_species
```

```python
# наблюдения с точностью до рода
df_genus = df[df['taxon_genus_name'].notnull() & df['taxon_species_name'].isnull()].copy()
df_genus.drop(subusual_columns, axis=1, inplace=True)

preusual_to_remove = preusual_columns.copy()
preusual_to_remove.remove('taxon_genus_name')

df_genus.drop(preusual_to_remove, axis=1, inplace=True)
df_genus.drop(['taxon_species_name'], axis = 1, inplace=True)
df_genus
```

```python
# наблюдения с точностью до подсемейства
df_subfamily = df[df['taxon_subfamily_name'].notnull() & df['taxon_genus_name'].isnull()].copy()

df_subfamily.drop(subusual_columns, axis=1, inplace=True)

preusual_to_remove = preusual_columns.copy()
preusual_to_remove.remove('taxon_subfamily_name')

df_subfamily.drop(preusual_to_remove, axis=1, inplace=True)
df_subfamily.drop(['taxon_species_name'], axis = 1, inplace=True)
df_subfamily
```
